```python
# Example: presidential heights
import pandas as pd
import numpy as np

data = pd.read_csv('president_heights.csv')

#print(data)
#print(data['height(cm)'])
heights = np.array(data['height(cm)'])
print(heights)

print(f"Minimum height: {heights.min()}")
print(f"Maximum height: {heights.max()}")
print(f"Mean height: {heights.mean()}")
print(f"Standard deviation in average height: {heights.std()}")

print(f"25% height: {np.percentile(heights, 25)}")
print(f"Median height: {np.median(heights)}")
print(f"75% height: {np.percentile(heights, 75)}")
```

```python
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

# histogrms are based on bins.  What you send is a list to be binned, and how many bins (if you want to control that)
plt.hist(heights)
# set 10 bins, then 5 bins
# plt.hist(heights, 5)
# flip the axis?  
# plt.hist(heights, orientation='horizontal')
plt.title('Height Distribution of US Presidents')
plt.xlabel('height (cm)')
plt.ylabel('number')
```
